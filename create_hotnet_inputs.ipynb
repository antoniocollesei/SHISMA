```python
import pandas as pd
import numpy as np
```

## create scores

```python
df = pd.read_csv("shap_values_CIK_cells_average.csv")
df = df.abs()
genes = df.columns

# create directory for scores
!mkdir shapelet_scores

for i in range(len(df)):
  row = df.iloc[i]
  if row.sum() != 0:
    row.T.to_csv(f"shapelet_scores/scores_row_{i}.csv", header=False, sep="\t")
```

## create edges

```python
# Create a gene-index file
gene2index = pd.Series(index=genes, data=range(1, len(genes) + 1))


# Load edges
edges = pd.read_csv("FIsInGene_122921_with_annotations.txt", sep="\t")

# Keep edges that are in genes
edges = edges[edges['Gene1'].isin(genes) & edges['Gene2'].isin(genes)]

# Remove edges with score < 1
edges = edges[edges['Score'] >= 1]
edges[['Gene1', 'Gene2']].to_csv("shapelet_scores/network.csv", header=False, sep="\t", index=False)

# Map genes to indexes
edges['Gene1'] = edges['Gene1'].map(gene2index)
edges['Gene2'] = edges['Gene2'].map(gene2index)

# Print to csv only Gene1, Gene2
edges[['Gene1', 'Gene2']].to_csv("shapelet_scores/network_edge_list.csv", header=False, sep="\t", index=False)

index2gene = pd.Series(data=genes, index=range(1, len(genes) + 1))
index2gene.to_csv("shapelet_scores/network_index_gene.csv", header=False, sep="\t")
```
