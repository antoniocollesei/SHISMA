```python
import pandas as pd
import numpy as np
```

## create scores

```python
df = pd.read_csv("shap_values/shap_values_CIK_cells_average.csv")
df = df.abs()
genes = df.columns

# create directory for scores
!mkdir shapelet_scores

for i in range(len(df)):
  row = df.iloc[i]
  if row.sum() != 0:
    row.T.to_csv(f"shapelet_scores/scores_row_{i}.csv", header=False, sep="\t")
```

## create edges

```python
# Create a gene-index file
gene2index = pd.Series(index=genes, data=range(1, len(genes) + 1))


# Load edges
edges = pd.read_csv("string_interactions_short.tsv", sep="\t")
network_name = "string"

# Keep edges that are in genes
edges = edges[edges['node1'].isin(genes) & edges['node2'].isin(genes)]

edges[['node1', 'node2']].to_csv("shapelet_scores/" + network_name, header=False, sep="\t", index=False)

# Map genes to indexes
edges['node1'] = edges['node1'].map(gene2index)
edges['node2'] = edges['node2'].map(gene2index)

# Print to csv only node1, node2
edges[['node1', 'node2']].to_csv("shapelet_scores/" + network_name + "_edge_list", header=False, sep="\t", index=False)

index2gene = pd.Series(data=genes, index=range(1, len(genes) + 1))
index2gene.to_csv("shapelet_scores/" + network_name + "_index_gene", header=False, sep="\t")
```
